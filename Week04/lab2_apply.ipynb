```python
import pandas as pd
pd.__version__
```

CSV is based on 
https://www.bowl.com/Open_Championships/Open_Championships_Home/Past_Results_and_History/

Load the file into Pandas

```python
data_url = "https://raw.githubusercontent.com/msaricaumbc/DS601_Fall21/main/Week04/data/bowling_stats.csv"

df = pd.read_csv(data_url)
```

```python
df
```

```python
df.columns = ["Year","City","State","Points1","Points2"]
df
```

```python
df.loc[df['Points2'].notnull(), 'Total_Points'] = df['Points1'] *1000 + df['Points2']
df.loc[df['Points2'].isnull(), 'Total_Points'] = df['Points1']
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
df
```

```python
df = df.drop(72)
```
