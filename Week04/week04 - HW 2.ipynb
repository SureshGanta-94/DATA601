```python
import pandas as pd
pd.__version__
```

```python
# Reading data from csv file
data_url = "https://raw.githubusercontent.com/suresh1308/Files/main/RollingSystemDemand_Data.csv"
df = pd.read_csv(data_url)
df
```

```python
# using pd.set_options to check full table
pd.set_option("display.max_rows", None, "display.max_columns", None)
# renaming header to desired values
header = {'HDR':'HDR', 'ROLLING SYSTEM DEMAND': 'ROLLING SYSTEM', 'Unnamed: 2':'Demand'}
df.rename(columns= header, inplace = True)
```

```python
df
```

```python
# Creating "Date & Time" column assigned formated "Rolling system" to it
s = df['ROLLING SYSTEM'].astype(str)
df['Date & Time'] = pd.to_datetime(s.str[:14],format ='%Y%m%d%H%M%S')
```

```python
df
```

```python
# Created dictonary to collect hourly data
dict = {}
for i in range(0,len(df['Demand']),12):
    total = df['Demand'].iloc[i:11+i].sum()
    dict[int(i/12)+1] = total
dict
```

```python
# Creating another dataframe for 10 days using above dictonary
df_for_10_days = pd.DataFrame.from_dict(dict,orient = 'index', columns = ['consumption'])
```

```python
# Added hours data to the above dataframe
import numpy as np
df_for_10_days['Hours'] = np.arange(len(df_for_10_days)) +1
df_for_10_days
```

```python
# Slicing dataframe to get 24 hours of data(for 1 day)
df_for_one_day = df_for_10_days[0:24]
```

```python
df_for_one_day
```

```python
# Plotting bar chart for 24 hours of data(for 1 day)
import matplotlib.pyplot as plt
df_for_one_day.plot.bar(x='Hours',y='consumption',figsize=(8,6))
plt.xlabel('Per Hour')
plt.ylabel('Consumption')
plt.title('Power consumption per hour for a day')
plt.show()
```

```python
# Plotting bar chart for 240 hours of data(for 10 days)
df_for_10_days.plot.bar(x='Hours',y='consumption',figsize=(30,8))
plt.xlabel('Per Hour')
plt.ylabel('Consumption')
plt.title('Power consumption per hour for 10 days')
plt.show()
```
