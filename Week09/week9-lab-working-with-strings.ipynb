__Your Task__

- In  `Q12_Part_4_TEXT` they ask the tools the participants are using day to day. However as you might observe the answers are not standardized. 

- Find all the answers with 'jupyter' mentioned in it.

- Find all the participant who are Python users.

- Find all R users. 

```python
df.head()
```

```python
import re
```

```python
#Q1
pattern = re.compile(r'j.*r',re.I)
```

```python
df['jupyter'] = [True if pattern.search(str(s)) != None else False for s in df['Q12_Part_4_TEXT']]
```

```python
#Q2
pattern2 = re.compile(r'\bpython\b',re.I)
```

```python
df['python'] = [True if pattern2.search(str(s)) != None else False for s in df['Q12_Part_4_TEXT']]
```

```python
#Q3
pattern3 = re.compile(r'\bR\b|R.?Studio',re.I)
```

```python
df['R'] = [True if pattern3.search(str(s)) != None else False for s in df['Q12_Part_4_TEXT']]
```
